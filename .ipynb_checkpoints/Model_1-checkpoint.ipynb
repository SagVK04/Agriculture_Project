```python
label_veg = LabelEncoder()
df['veg_encoded'] = label_veg.fit_transform(df['vegetable'])
df.head(5)
```

```python
label_weather = LabelEncoder()
df['weather_encoded'] = label_weather.fit_transform(df['weather_condition'])
df.head(5)
```

```python
label_demand = LabelEncoder()
df['demand_encoded'] = label_demand.fit_transform(df['demand_level'])
df.head(5)
```

```python
numeric_df = df.select_dtypes(include=['number'])
numeric_df.head(10)
```

```python
numeric_df.columns.tolist()
```

```python
sample_corr = numeric_df.corr(numeric_only = True)
sample_corr['daily_price'].sort_values(ascending = False)
```

```python
selected_df = numeric_df[['weather_encoded',
    'veg_encoded',
    'demand_encoded',
    'daily_price']]
selected_df.head(7)
```

```python
selected_df.describe()
```

```python
df[['demand_level', 'vegetable','weather_condition','daily_price']].sample(10)
```

```python
selected_df.columns.tolist()
```

```python
x = selected_df[['weather_encoded', 'veg_encoded', 'demand_encoded']]
x
```

```python
y = selected_df['daily_price']
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
print("Mean Absolute Error(Linear Regression): ",mean_absolute_error(y_pred,y_test))
```

```python
rf = RandomForestRegressor(
    n_estimators=200,
    max_depth=12,
    random_state=42
)
rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)
print("Random Forest MAE:", mean_absolute_error(y_test, y_pred))
```

```python
def user_input():
    import builtins
    input = builtins.input
    
    print("\nEnter details for price prediction:\n")
        
    veg = builtins.input("Vegetable (Potato/Onion/Tomato/Cauliflower/Cabbage/Brinjal): ")
    temp = float(builtins.input("Temperature (C): "))
    rain = float(builtins.input("Rainfall (mm): "))
    humidity = float(builtins.input("Humidity (%): "))
    weather = builtins.input("Weather Condition (Sunny/Rainy/Heatwave/Cold Wave/Heavy Rain): ")
    demand = builtins.input("Demand Level (Low/Medium/High): ")
        
        
    weather_in = label_weather.transform([weather])[0]
    veg_in = label_veg.transform([veg])[0]
    demand_in = label_demand.transform([demand])[0]
    data = pd.DataFrame([{
        "weather_encoded": weather_in,	
        "veg_encoded": veg_in,	
        "demand_encoded": demand_in
    }])
    print("Predicted Price(Rs. Per Kg): ",round(rf.predict(data)[0],2))
```
